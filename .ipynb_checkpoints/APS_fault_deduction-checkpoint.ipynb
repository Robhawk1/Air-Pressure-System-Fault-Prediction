```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
import pandas as pd

# Read the training data
train_df = pd.read_csv("/home/rob_hawk/Downloads/archive/aps_failure_training_set.csv", na_values="na")

# Read the testing data
test_df = pd.read_csv("/home/rob_hawk/Downloads/archive/aps_failure_test_set.csv", na_values="na")


# Display the first 5 rows of the data
print(train_df.head())
print(test_df.head())
```

```python
#Changing the target class values to integers
train_df['class'] = train_df['class'].map({'neg':0, 'pos':1})
test_df['class'] = test_df['class'].map({'neg':0, 'pos':1})

```

```python
# Plotting Missing values count for each column
missing = train_df.isna().mean()
missing = missing[missing > 0]
missing.sort_values(inplace=True, assending = False)
missing.plot.bar(figsize=(15,5))
plt.ylabel("% of Missing values")
plt.show()
```

```python
#Dropping column with more than 70% missing values
missing = train_df.isna().mean()
missing = missing[missing > 0.7]
train_df.drop(missing.index, axis=1, inplace=True)
```

```python
#Dropping column with more than 70% missing values
missing = test_df.isna().mean()
missing = missing[missing > 0.7]
test_df.drop(missing.index, axis=1, inplace=True)
```

```python
#Separating X and Y from datasets
train_df_X = train_df.loc[:,train_df.columns != 'class']
train_df_Y = train_df.loc[:,train_df.columns == 'class']
test_df_X = test_df.loc[:,test_df.columns != 'class']
test_df_Y = test_df.loc[:,test_df.columns == 'class']
```

```python
train_df_X = train_df_X.apply(pd.to_numeric)

train_df_X= train_df_X.fillna(train_df_X.mean()).dropna(axis =1 , how ='all')

scaler = StandardScaler()

scaler.fit(train_df_X)

train_df_X = scaler.transform(train_df_X)

pca = PCA(0.95)

pca.fit(train_df_X)

pca.n_components_
```

```python
train_df_X = pca.transform(train_df_X)
train_df_X= pd.DataFrame(train_df_X)

```

```python
train_df_X.shape
```

```python
#Doing same process for test data
test_df_X = test_df_X.apply(pd.to_numeric)
test_df_X = test_df_X.fillna(test_df_X.mean()).dropna(axis =1 , how ='all')
scaler = StandardScaler()
scaler.fit(test_df_X)
test_df_X = scaler.transform(test_df_X)

test_df_X = pca.transform(test_df_X)
test_df_X= pd.DataFrame(test_df_X)
```

```python
test_df_X.shape
```

```python
# Splitting the train dataset into training and validation set
X_train, X_val, Y_train, Y_val = train_test_split(train_df_X, train_df_Y, test_size=0.2, random_state=42)

# apply SMOTE to the training data
smote = SMOTE(sampling_strategy='auto')
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

```python
# Checking the class imbalance after smote
plt.figure(figsize=(15,8))
sns.distplot(Y_train_smote);
```

```python
#function to train the model
def train_model(model, X_train, y_train):
    model.fit(X_train, y_train)
    return model
```

```python
#function to evaluate the model's performance
def evaluate_model(model, X_val, y_val):
    y_val_pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, y_val_pred)
    precision = precision_score(y_val, y_val_pred)
    recall = recall_score(y_val, y_val_pred)
    f1 = f1_score(y_val, y_val_pred)
    return accuracy, precision, recall, f1
```

```python
#function to print the performance scores
def print_scores(scores, model_name):
    print(f'{model_name} :')
    print("Accuracy: ", scores[0])
    print("Precision: ", scores[1])
    print("Recall: ", scores[2])
    print("F1 Score: ", scores[3])
```

```python
#initialize models
log_reg = LogisticRegression(C =0.0001,penalty = 'l2')
svm = SVC(C=0.25, kernel='rbf', degree=3, gamma='auto', random_state=42)
rf = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=2, min_samples_leaf=1, max_features='sqrt', random_state=42)
xgb = XGBClassifier(n_estimators=100, max_depth=3, learning_rate=0.1, subsample=0.8, colsample_bytree=0.8, reg_alpha=0.01, reg_lambda=1, random_state=42)
```

```python
#train models
trained_log_reg = train_model(log_reg, X_train_smote, Y_train_smote.squeeze())
trained_svm = train_model(svm, X_train_smote, Y_train_smote.squeeze())
trained_rf = train_model(rf, X_train_smote, Y_train_smote.squeeze())
trained_xgb = train_model(xgb, X_train_smote, Y_train_smote.squeeze())
```

```python
# Get predictions on test set
log_reg_predictions = trained_log_reg.predict(test_df_X)
svm_predictions = trained_svm.predict(test_df_X)
rf_predictions = trained_rf.predict(test_df_X)
xgb_predictions = trained_xgb.predict(test_df_X)

# Evaluate predictions
log_reg_scores = evaluate_model(trained_log_reg, test_df_X, test_df_Y)
svm_scores = evaluate_model(trained_svm, test_df_X, test_df_Y)
rf_scores = evaluate_model(trained_rf, test_df_X, test_df_Y)
xgb_scores = evaluate_model(trained_xgb, test_df_X, test_df_Y)

# Print scores
print_scores(log_reg_scores, 'Logistic Regression')
print_scores(svm_scores, 'SVM')
print_scores(rf_scores, 'Random Forest')
print_scores(xgb_scores, 'XGBoost')
```
